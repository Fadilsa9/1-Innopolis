## основные термины при работе с pandas

Pandas вводит ряд новых слов которые будут использованы при работе с ней. Имеет смысл с ними ознакомиться.

*__Series__* (серия)- это одномерный, именованный (помеченный) массив إنها مصفوفة أحادية البعد، مسماة (مُسمَّاة), который может хранить данные любого типа, включая целые числа, числа с плавающей точкой, строки и объекты Python.

Базовая структура серии состоит из двух компонентов: данных и индекса.

В серии *__данные__* представляют собой одномерный массив NumPy, содержащий фактические значения данных.

```python
# just an example of series(here we will see lables and names)
data_frame_covid.T.iloc[:,0]

# explaining  ==>  https://chat.hix.ai/ar/share/cm0g2ntwx0gey5j0qaog0rkhq
# questions   ==>  https://chat.hix.ai/ar/share/cm0g2ze600gtjlze5nspwrude
```

```python
# Question 1:
# Given the following DataFrame that contains information about different fruits:


import pandas as pd

data = {
    'Fruit': ['Apple', 'Banana', 'Cherry'],
    'Color': ['Red', 'Yellow', 'Red'],
    'Price': [1.2, 0.5, 2.0]
}
df_fruits = pd.DataFrame(data)

# data

### What is the value in the first row of the 'Color' column after transposing the DataFrame?
first_color_value = df_fruits.T.iloc[1, 0]  # Accessing the first element of the 'Color' row
print(first_color_value)
```

```python
# find the type of series
type(data_frame_covid.T.iloc[:,0])    # == pandas.core.series.Series
```

*__Индекс__* - это структура, похожая на массив, которая помечает каждый элемент массива данных الذي يمثل كل عنصر من عناصر مجموعة البيانات. По умолчанию индекс представляет собой последовательность целых чисел, начинающуюся с нуля, но вы также можете предоставить пользовательские метки (Метка = label 🏷️ العلامة = التسمية) для индекса. Это облегчает обращение и манипулирование элементами данных на основе их меток.

Чтобы создать серию, вы можете использовать конструктор pd.Series () и передать список, словарь или массив NumPy в качестве данных. Вы также можете передать необязательный параметр индекса, чтобы указать пользовательские метки для элементов данных.

*__DataFrame__* (ДатаФрейм, Таблица данных) - это двумерная, изменяемая в размере и гетерогенная структура табличных данных с маркированными осями (строками и столбцами). Его можно рассматривать как коллекцию объектов Series, где каждый столбец представляет собой объект Series. DataFrames предназначены для обработки широкого спектра типов данных, что делает его универсальным и мощным инструментом для обработки и анализа данных. Базовая структура DataFrame состоит из трех компонентов: данных, индекса и столбцов.

В DataFrame данные хранятся в виде коллекции из одного или нескольких объектов Series, каждый столбец которых представляет собой отдельную Series. Данные в каждой Series могут быть различных типов, таких как целые числа, числа с плавающей точкой или строки.

Как Series так и DataFrame предоставляют методы для анализа данных, такие как:
- Выборка данных: можно указать особые строки и колонки с которыми вы будете работать, используя индексы,слайсы,условия
- Очистка данных: можно заполнять пропущенные данные, удалять записи в которых не хватает данных.
- Преобразование данных: можно применять функции по всей длинне фрейма или по его отдельным частям

- Сортировка: можно упорядочить данные основываясь на значениях в колонках, а также по своим критериям
- Функциональность временных рядов: вы можете работать с данными временных рядов, пересчитывать данные с разной частотой и выполнять вычисления с использованием скользящего окна в DataFrame или Series.

# Summary

Numpy это основополагающая библиотека для анализа данных и алгоритмов машинного обучения на python. Она имеет свои преимущества и свои недостатки относительно стандартной реализации списков в cpython. Для numpy реализованы некоторые методы линейнной алгебры и обраотки сигналов, а в scipy, основанной на numpy реализованы еще и дополнительные алгоритмы по оптимизации и решению дифуров.
При работе с прямоугольными именованными таблицами данных классическим инструментом можно считать pandas. Pandas представляет собой библиотеку предназначенную для обработки датафреймов - именованных прямоугольных таблиц. Эти таблицы состоят из серий данных, колонок и индексов. Последние помогают находить строки в датафрейме. Основными вариантами использования pandas являются обработка, фильтрация, трфнсформирование и очистка данных. В отдельных случаях возможна работа с геопространственными данными через дополнение geopandas, однако геопространственные термины и типы данных скорее являются отдельной темой для лекции.
