```python
# Question 1:
# Given the following DataFrame that contains information about different fruits:


import pandas as pd

data = {
    'Fruit': ['Apple', 'Banana', 'Cherry'],
    'Color': ['Red', 'Yellow', 'Red'],
    'Price': [1.2, 0.5, 2.0]
}
df_fruits = pd.DataFrame(data)

# data

### What is the value in the first row of the 'Color' column after transposing the DataFrame?
first_color_value = df_fruits.T.iloc[1, 0]  # Accessing the first element of the 'Color' row
print(first_color_value)
```

```python
# find the type of series
type(data_frame_covid.T.iloc[:,0])    # == pandas.core.series.Series
```
