## Анализ тональности

**Анализ тональности** (Sentiment Analysis) — это процесс определения
эмоциональной окраски текста. Чаще всего анализируют, является ли текст
положительным, отрицательным или нейтральным. Данный метод широко
используется в маркетинге, обслуживании клиентов и исследованиях
общественного мнения, чтобы понять, как потребители воспринимают
продукты, бренды или события.

#### Зачем нужен анализ тональности?


* **Мониторинг бренда**: Определение общественного мнения о компании или продукте. Компании могут отслеживать тональность обсуждений в социальных сетях, блогах и новостях, чтобы понять, как воспринимается их бренд и как меняются настроения аудитории со временем.

* **Исследования рынка**: Анализ отзывов и комментариев клиентов помогает компаниям выявлять ключевые проблемы и сильные стороны продукта. Это позволяет лучше понимать потребности клиентов и оперативно адаптироваться к их ожиданиям.

* **Автоматизация поддержки**: Быстрое определение негативных отзывов для оперативного реагирования. Системы анализа тональности помогают службам поддержки выделять негативные отзывы или жалобы, чтобы приоритетно работать с проблемными клиентами и предотвращать эскалацию конфликтов.

* **Оценка маркетинговых кампаний**: Компании могут измерять реакцию аудитории на рекламные кампании или запуск новых продуктов. Анализ тональности позволяет понять, как восприняты те или иные маркетинговые сообщения и корректировать стратегию в реальном времени.



#### Подходы к анализу тональности

* **Правила и словари**: Используют предопределенные списки слов с присвоенными им тональностями (позитивные, негативные или нейтральные). Этот подход основывается на поиске в тексте слов из словаря и присвоении тональности тексту в зависимости от встреченных слов. Хотя такие системы просты и быстры в реализации, они могут плохо справляться с учетом контекста и многозначных слов, а также не всегда адаптивны к изменениям языка и новым терминам.

* **Машинное обучение**: Обучение моделей на размеченных данных для определения тональности текста. Этот подход включает использование методов классического машинного обучения, таких как логистическая регрессия, случайный лес, SVM и другие. Тексты преобразуются в числовые признаки (например, с помощью TF-IDF или мешка слов), на которых обучается модель для предсказания тональности. Машинное обучение позволяет строить более гибкие модели, но для достижения хороших результатов требуется достаточное количество размеченных данных.

* **Глубокое обучение и трансформеры**: Использование сложных моделей для учета контекста и семантики. Современные методы анализа тональности основаны на нейронных сетях, включая рекуррентные сети (LSTM, GRU) и трансформеры (BERT, GPT). Эти модели могут учитывать контекст предложения и определять тональность на более глубоком уровне. Они способны улавливать сложные зависимости между словами и справляться с задачами анализа длинных текстов и предложений с изменяющимся контекстом. Модели трансформеров особенно эффективны благодаря своей способности захватывать глобальные зависимости в тексте, что делает их стандартом для многих задач NLP.


**VADER** (Valence Aware Dictionary and sEntiment Reasoner) — это алгоритм для анализа тональности текстов, специально разработанный для обработки коротких текстов, таких как сообщения в социальных сетях, отзывы и комментарии. VADER использует комбинацию словарного подхода и правил для определения эмоциональной окраски текста.

### Пример

```python
# Импортируем необходимые библиотеки
from nltk.sentiment.vader import SentimentIntensityAnalyzer  # Импорт анализатора настроений из NLTK
import nltk  # Импорт NLTK для загрузки лексикона VADER
nltk.download('vader_lexicon')  # Загрузка лексикона VADER из NLTK

# Создаем экземпляр SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()

# Задаем текст для анализа
text = "This is the best day ever!"

# Вычисляем оценку настроений с помощью VADER
score = sid.polarity_scores(text)

# Выводим результаты
print(score)
```



- **`'neg'`: 0.0** — это доля негативной тональности в тексте. В данном случае значение 0.0 указывает на то, что в тексте нет негативных эмоций.

- **`'neu'`: 0.527** — это доля нейтральной тональности. Значение 0.527 (или 52.7%) говорит о том, что чуть больше половины текста выражает нейтральные эмоции, не имея явной позитивной или негативной окраски.

- **`'pos'`: 0.473** — это доля позитивной тональности. Значение 0.473 (или 47.3%) означает, что почти половина текста имеет позитивный заряд.

- **`'compound'`: 0.6696** — это композитная оценка, которая является сводной метрикой, отражающей общую тональность текста. Значение композитной оценки колеблется от -1 (полностью негативный) до +1 (полностью позитивный). В данном случае, значение 0.6696 указывает на то, что текст имеет в целом позитивную тональность.

##### Интерпретация
Этот результат указывает на то, что анализируемый текст в основном нейтральный с небольшим уклоном в позитивную сторону, так как доля позитивной тональности почти равна нейтральной. Общая композитная оценка также подтверждает это, показывая, что текст имеет положительное настроение.

```python
# Задаем текст для анализа
text = "This is the bad day ever!"

# Вычисляем оценку настроений с помощью VADER
score = sid.polarity_scores(text)

# Выводим результаты
print(score)
```

### Обучение модели

```python
# Импортируем необходимые модули из библиотеки NLTK для обработки текста
from nltk.corpus import stopwords  # Для работы со списками стоп-слов
from nltk.stem import WordNetLemmatizer  # Для лемматизации слов

# Импортируем TfidfVectorizer для преобразования текстов в числовые векторы
from sklearn.feature_extraction.text import TfidfVectorizer

# Импортируем функции для разделения данных и моделирования
from sklearn.model_selection import train_test_split  # Для разделения данных на обучающую и тестовую выборки
from sklearn.ensemble import RandomForestClassifier  # Классификатор случайного леса

# Импортируем метрики для оценки моделей
from sklearn.metrics import classification_report, accuracy_score, roc_auc_score  # Для оценки качества модели
```

```python
import re

def clean_text(text):
    text = re.sub(r'\W', ' ', text)  # Заменяем все, что не является буквой или цифрой, на пробел
    text = re.sub(r'\s+', ' ', text)  # Убираем лишние пробелы
    return text.lower()  # Приводим текст к нижнему регистру

data = pd.read_csv('drive/MyDrive/train 2.csv')

# Оставим только 2 класса для предсказания
data = data[data['sentiment'].isin(['positive', 'negative'])]

data['clean_text'] = data['text'].apply(clean_text)

data.head(3)
```

```python
nltk.download('stopwords')
```

```python
nltk.download('wordnet')  # Загружаем словарь WordNet для лемматизации

# Создаем множество стоп-слов на английском языке
stop_words = set(stopwords.words('english'))

# Инициализируем лемматизатор
lemmatizer = WordNetLemmatizer()

# Функция для лемматизации текста и удаления стоп-слов
def preprocess_text(text):
    words = text.split()  # Разделяем текст на слова
    # Лемматизируем каждое слово и удаляем стоп-слова
    words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]
    return ' '.join(words)  # Объединяем слова обратно в строку

# Применение функции предобработки к очищенному тексту
data['processed_text'] = data['clean_text'].apply(preprocess_text)
```

```python
# Разделение данных на обучающие и тестовые наборы
X_train, X_test, y_train, y_test = train_test_split(
    data['processed_text'],  # Признаки (тексты)
    data['sentiment'],       # Целевая переменная (тональность)
    test_size=0.2,          # 20% данных будут отведены под тестовый набор
    random_state=42          # Устанавливаем сид для воспроизводимости
)

# Векторизация текста
vectorizer = TfidfVectorizer(max_features=1000)  # Создаем объект TfidfVectorizer с ограничением на 1000 признаков
X_train_tfidf = vectorizer.fit_transform(X_train).toarray()  # Обучаем TF-IDF на обучающих данных и преобразуем в массив
X_test_tfidf = vectorizer.transform(X_test).toarray()  # Применяем TF-IDF к тестовым данным и преобразуем в массив
```

```python
from sklearn.ensemble import RandomForestClassifier  # Импортируем классификатор случайного леса
from sklearn.linear_model import LogisticRegression  # Импортируем логистическую регрессию
from xgboost import XGBClassifier  # Импортируем классификатор XGBoost
from lightgbm import LGBMClassifier  # Импортируем LightGBM классификатор

# Создаем словарь моделей для удобного обращения
models = {
    'Random Forest': RandomForestClassifier(),  # Классификатор случайного леса
    'Logistic Regression': LogisticRegression(max_iter=1000),  # Логистическая регрессия с максимальным количеством итераций 1000
    'LightGBM': LGBMClassifier(),  # Классификатор LightGBM
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='logloss')  # Классификатор XGBoost с отключенным кодировщиком меток и метрикой логарифмической потери
}
```

```python
# Кодируем метки классов в числовой формат
y_train_encoded = y_train.map({'positive': 1, 'negative': 0})  # Преобразуем 'positive' в 1, а 'negative' в 0 для обучающего набора
y_test_encoded = y_test.map({'positive': 1, 'negative': 0})    # Аналогично для тестового набора

# Обучение моделей и расчет метрик
results = {}  # Словарь для хранения результатов

# Проходим по каждой модели из словаря
for model_name, model in models.items():
    model.fit(X_train_tfidf, y_train_encoded)  # Обучаем модель на обучающих данных
    y_proba = model.predict_proba(X_test_tfidf)[:, 1]  # Получаем вероятности положительного класса для тестовых данных
    auc = roc_auc_score(y_test_encoded, y_proba)  # Вычисляем ROC-AUC Score
    results[model_name] = auc  # Сохраняем результат в словарь
    print(f'{model_name} - ROC-AUC Score: {auc:.4f}')  # Выводим название модели и её ROC-AUC Score
```

```python
from sklearn.metrics import roc_auc_score, roc_curve  # Импортируем необходимые функции для вычисления ROC-кривой
import matplotlib.pyplot as plt  # Импортируем библиотеку для визуализации

# Настраиваем размер графика
plt.figure(figsize=(10, 8))

# Проходим по каждой модели из словаря
for model_name, model in models.items():
    # Получаем вероятности положительного класса для тестовых данных
    y_proba = model.predict_proba(X_test_tfidf)[:, 1]

    # Вычисляем значения для ROC-кривой
    fpr, tpr, _ = roc_curve(y_test.map({'positive': 1, 'negative': 0}), y_proba)

    # Строим график для каждой модели
    plt.plot(fpr, tpr, label=f'{model_name} (AUC = {results[model_name]:.2f})')

# Добавляем диагональную линию (линия случайного угадывания)
plt.plot([0, 1], [0, 1], 'k--')  # 'k--' - черная пунктирная линия

# Устанавливаем пределы осей
plt.xlim([0.0, 1.0])  # Ограничиваем ось X от 0 до 1
plt.ylim([0.0, 1.05])  # Ограничиваем ось Y от 0 до 1.05

# Добавляем подписи к осям и заголовок
plt.xlabel('False Positive Rate')  # Подпись оси X
plt.ylabel('True Positive Rate')    # Подпись оси Y
plt.title('ROC Curve for Different Models')  # Заголовок графика

# Добавляем легенду в нижний правый угол
plt.legend(loc='lower right')

# Отображаем график
plt.show()
```

## Возвращаемся в векторные представления слов

Векторные представления слов (эмбеддинги) позволяют представить слова в
виде числовых векторов, что упрощает обработку текстов моделями
машинного обучения. Основная идея заключается в том, что семантически
похожие слова имеют схожие векторные представления

### Зачем нужны векторные представления?

**Сокращение размерности**: Эмбеддинги преобразуют слова из огромного дискретного
пространства в компактное непрерывное пространство меньшей размерности.

**Учет контекста**: Векторные представления могут захватывать контекстные
особенности слов, что делает их полезными для сложных NLP-задач.

**Устойчивость к шуму**: Эмбеддинги менее подвержены шумикам и позволяют моделям
быть более устойчивыми к изменению текстов.

## Традиционные методы векторизации

### Мешок слов (Bag of Words, BoW)

Модель "мешок слов" представляет текст как вектор, где каждый элемент соответствует частоте встречаемости слова в тексте.

* Плюсы: Простота реализации и понимания.
* Минусы: Игнорирует порядок слов и контекст; векторизация на уровне
документов может приводить к высоким разреженным матрицам.

#### Пример

```python
from sklearn.feature_extraction.text import CountVectorizer  # Импортируем CountVectorizer для создания мешка слов

# Пример текстов для векторизации
texts = ["собака лает", "кошка мяукает", "собака и кошка"]

# Создаем объект CountVectorizer для преобразования текста в матрицу частот слов
vectorizer = CountVectorizer()

# Обучаем векторизатор на текстах и преобразуем их в матрицу
X = vectorizer.fit_transform(texts)

# Выводим мешок слов в виде массива
print("Мешок слов:\n", X.toarray())

# Выводим список всех уникальных слов (фич) из текстов
print("Фичи:", vectorizer.get_feature_names_out())
```

### TF-IDF (Term Frequency-Inverse Document Frequency)

TF-IDF улучшает BoW, взвешивая частоту слов с учетом их распространенности в
других документах.

* Плюсы: Учитывает редкие слова, уменьшает влияние часто встречающихся слов.
* Минусы: Все еще не учитывает порядок слов и контекст

#### Пример

```python
from sklearn.feature_extraction.text import TfidfVectorizer  # Импортируем TfidfVectorizer для вычисления TF-IDF

# Пример текстов для векторизации
texts = ["собака лает", "кошка мяукает", "собака и кошка"]

# Создаем объект TfidfVectorizer для преобразования текстов в TF-IDF матрицу
vectorizer = TfidfVectorizer()

# Обучаем векторизатор на текстах и преобразуем их в TF-IDF матрицу
X = vectorizer.fit_transform(texts)

# Выводим TF-IDF в виде массива
print("TF-IDF:\n", X.toarray())

# Выводим список всех уникальных слов (фич) из текстов
print("Фичи:", vectorizer.get_feature_names_out())
```

### Word2Vec

**Word2Vec** — это алгоритм, разработанный в 2013 году командой исследователей под
руководством Томаша Миколова в Google. Этот метод преобразует слова в числовые
векторы, которые захватывают семантические отношения между словами. Идея
заключается в том, чтобы моделировать слова в многомерном пространстве, где слова
с похожими значениями находятся ближе друг к другу.

image.png

### Принцип работы Word2Vec

Word2Vec обучает нейронную сеть на огромных объемах текста и использует два
основных подхода:
1. CBOW (Continuous Bag of Words): предсказывает текущее слово, используя
окружающие его слова (контекст).
2. Skip-gram: предсказывает окружающие слова (контекст), используя текущее слово.

Модель, предложенная Миколовым очень проста (и потому так хороша) — мы будем
предсказывать вероятность слова по его окружению (контексту). То есть мы будем
учить такие вектора слов, чтобы вероятность, присваиваемая моделью слову была
близка к вероятности встретить это слово в этом окружении в реальном тексте.

Процесс тренировки устроен следующим
образом: мы берем последовательно (2k+1)
слов, слово в центре является тем словом,
которое должно быть предсказано. А
окружающие слова являются контекстом
длины по k с каждой стороны. Каждому
слову в нашей модели сопоставлен
уникальный вектор, который мы меняем в
процессе обучения нашей модели.

В целом, этот подход называется CBOW —
continuous bag of words, continuous потому,
что мы скармливаем нашей модели
последовательно наборы слов из текста, a
BoW потому что порядок слов в контексте не
важен.

Также Миколовым сразу был предложен
другой подход — прямо противоположный
CBOW, который он назвал skip-gram, то есть
“словосочетание с пропуском”. Мы пытаемся
из данного нам слова угадать его контекст
(точнее вектор контекста). В остальном
модель не претерпевает изменений.

image.png

Что стоит отметить: хотя в модель не заложено явно
никакой семантики, а только статистические свойства
корпусов текстов, оказывается, что натренированная
модель word2vec может улавливать некоторые
семантические свойства слов. Классический пример
из работы автора:

Слово "мужчина" относится к слову "женщина" так
же, как слово "дядя" к слову "тётя", что для нас
совершенно естественно и понятно, но в других
моделям добиться такого же соотношения векторов
можно только с помощью специальных ухищрений.
Здесь же — это происходит естественно из самого
корпуса текстов. Кстати, помимо семантических
связей, улавливаются и синтаксические, справа
показано соотношение единственного и
множественного числа.

### 2. Моделирование

__Основные параметры:__

* данные должны быть итерируемым объектом
* `size` — размер вектора,
* `window` — размер окна наблюдения,
* `min_count` — мин. частотность слова в корпусе,
* `sg` — используемый алгоритм обучения (0 — CBOW, 1 — Skip-gram),
* `sample` — порог для downsampling'a высокочастотных слов,
* `workers` — количество потоков,
* `alpha` — learning rate,
* `iter` — количество итераций,
* `max_vocab_size` — позволяет выставить ограничение по памяти при создании словаря (т.е. если ограничение привышается, то низкочастотные слова будут выбрасываться). Для сравнения: 10 млн слов = 1Гб RAM.

```python
%%time
# Эта команда используется в Jupyter Notebook для измерения времени выполнения ячейки кода

from gensim.models.word2vec import Word2Vec

# vector_size - размерность векторов, которые мы хотим обучить
# window - ширина окна контекста
# min_count - если слово встречается реже, для него не учим модель
model = Word2Vec(vector_size=100, window=5, min_count=3, workers=4)

# строительство словаря, чтобы обучение шло быстрее
model.build_vocab(sents_tokenize)

# обучение модели
# первый аргумент - наша выборка, генератор будет вкидывать в модель наши тексты, пока они не кончатся
# второй аргумент - число примеров в выборке
# третий аргумент - количество эпох обучения: сколько раз модель пройдётся по всему корпусу текстов
model.train(sents_tokenize, total_examples=model.corpus_count, epochs=100)
```
